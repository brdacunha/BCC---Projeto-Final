1) Faça o gráfico da Renda dos jogos do campeonato brasileiro de 2018 em função do público de cada jogo. Calcule a correlação entre essas grandezas.

```python
import matplotlib.pyplot as plt # importando a biblioteca para plotar o gráfico
import pandas as pd # para trabalhar com a base de dados

Brasileirão = pd.read_csv('https://raw.githubusercontent.com/BMClab/BasesComputacionais/master/dados/tabelaBrasileirao2018.csv')
Brasileirão
```

```python
plt.figure() 
plt.plot(Brasileirão['Público'], Brasileirão['Renda (R$)'], 
         marker='o', color='red', linestyle='') # plotando o gráfico da Renda em função do público.
plt.xlabel('Público', fontsize=13)
plt.ylabel('Renda (milhões (R$))', fontsize=13)
plt.show()
corr_renda_público = Brasileirão['Público'].corr(Brasileirão['Renda (R$)']) # calculando a correlação entre as colunas
print(f'A correlação entre Renda e Público é de {corr_renda_público:.2f}.') # renda e público
```
